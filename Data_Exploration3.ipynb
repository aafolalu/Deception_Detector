```python
import pandas as pd
```

```python
# CONVERTED FILE TYPE FROM TXT TO CSV
# LOAD DATA
review = pd.read_csv('/Users/adebisiafolalu/Desktop/OneDrive/Documents/MSc_Assignments/Semester2/NLP/assignment1/Deception_Detector/amazon_reviews.csv')
review.head()
```

```python
# PRINTING POSITVE FOR ALL RATING 4 & 5
# PRINTING NEGATIVE FOR ALL RATING 1 & 2
# REMOVING ALL RATING 3 REVIEWS
# REMOVING ALL RATING 5 REVIEWS EXCEPT THE FIRST 2949 RATING 5 REVIEWS
rating = review['RATING']

P, N, R = 0,0,0
Tot1, Tot2, Tot3 = 0,0,0
indexN = 0
indexA = []


for r in rating:
    if r  > 3:
        print("POSITIVE")
        P += 1
        Tot1 += P
        if Tot1 > 2949:
            indexA.append(indexN)
            
    elif r < 3:
        print("NEGATIVE")
        N += 1
        Tot2 += N
    else:
        indexA.append(indexN)
        R += 1
        Tot3 += R
    P, N, R = 0,0,0
    indexN += 1

       
review1 = review.drop(review.index[indexA])

print(Tot1)
print(Tot2)
print(Tot3)
print(len(review1))
review1.head(n=indexN)
    
```

```python
# SELECTING FIELD INPUTS
# sentiment
positive = review['RATING'] > 3
negative = review['RATING'] < 3

# labels
fake = review['LABEL'] == '__label1__'
real = review['LABEL'] == '__label2__'

# verified purchase
y = review['VERIFIED_PURCHASE'] == "Y"
n = review['VERIFIED_PURCHASE'] == "N"

pc = review['PRODUCT_CATEGORY'] == 'PC'
Wireless = review['PRODUCT_CATEGORY'] == 'Wireless'
Baby = review['PRODUCT_CATEGORY'] == 'Baby'
Beauty = review['PRODUCT_CATEGORY'] == 'Beauty'
```

```python
# CORRELATION BETWEEN SENTIMENT AND LABELS
print("Correlation between sentiment and labels")
print("Real positive sentiment vs fake label:  %f"%(positive.corr(fake)))
print("Real positive sentiment vs real label:  %f"%(positive.corr(real)))
print("Real negative sentiment vs fake label:  %f"%(negative.corr(fake)))
print("Real negative sentiment vs real label:  %f"%(negative.corr(real)))
```

```python
# CORRELATION BETWEEN SENTIMENT AND VERIFIED PURCHASES
print("Correlation between sentiment and verified purchases")
print("Real positive sentiment vs verified purchase:  %f"%(positive.corr(y)))
print("Real positive sentiment vs not verified purchase:  %f"%(positive.corr(n)))
print("Real negative sentiment vs verified purchase:  %f"%(negative.corr(y)))
print("Real negative sentiment vs not verified purchasel:  %f"%(negative.corr(n)))
```

```python
# CORRELATION BETWEEN SENTIMENT AND PRODUCT CATEGORY
print("Correlation between sentiment and product categories")
print("Real positive sentiment vs PC category:  %f"%(positive.corr(pc)))
print("Real positive sentiment vs Wireless category:  %f"%(positive.corr(Wireless)))
print("Real negative sentiment vs Baby categoryl:  %f"%(negative.corr(Baby)))
print("Real negative sentiment vs Beauty category:  %f"%(negative.corr(Beauty)))
```
